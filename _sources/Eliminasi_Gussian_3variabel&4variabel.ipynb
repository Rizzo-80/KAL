##Eliminasi Gussian
Eliminasi Gauss adalah prosedur dengan Operasi Baris Elementer untuk mereduksi sistem persamaan linear menjadi bentuk eselon baris yang mudah dipecahkan.Disini saya akan membuat contoh soal yang menggunakan eliminasi Gussian 3 variabel dan 4 variabel:

\
\begin{bmatrix}
1 & 2 & 1 & 8 \\
2 & 1 & -1 & 1 \\
-1 & 1 & 2 & 5
\end{bmatrix}




```
B = matrix(QQ, [[1, 2, 1, 8],
                [2, 1, -1, 1],
                [-1, 1, 2, 5]])

M = B

# Baris 1 - 2*Baris 0
M.add_multiple_of_row(1, 0, -2)

# Baris 2 + 1*Baris 0
M.add_multiple_of_row(2, 0, 1)

# Kalikan Baris 1 dengan -1/3
M.rescale_row(1, -1/3)

# Baris 2 - 3*Baris 1
M.add_multiple_of_row(2, 1, -3)

print("Hasil:")
print(M)
```

## Langkah 1: Eliminasi Kolom Pertama

Operasi baris:

R2 = R2 - 2R1  
R3 = R3 + R1  

Hasil:

$$
\begin{bmatrix}
1 & 2 & 1 & 8 \\
0 & -3 & -3 & -15 \\
0 & 3 & 3 & 13
\end{bmatrix}
$$


## Langkah 2: Membuat Pivot Menjadi 1

Operasi baris:

R2 = (-1/3) R2  

Hasil:

$$
\begin{bmatrix}
1 & 2 & 1 & 8 \\
0 & 1 & 1 & 5 \\
0 & 3 & 3 & 13
\end{bmatrix}
$$


## Langkah 3: Eliminasi Kolom Kedua

Operasi baris:

R3 = R3 - 3R2  

Hasil (Bentuk Eselon Baris):

$$
\begin{bmatrix}
1 & 2 & 1 & 8 \\
0 & 1 & 1 & 5 \\
0 & 0 & 0 & -2
\end{bmatrix}
$$


## Kesimpulan

Baris terakhir menghasilkan:

0 = -2  

Karena ini tidak mungkin benar, maka sistem **tidak memiliki solusi (inkonsisten)**.

```
# Matriks Augmentasi Awal
B = matrix(QQ, [
    [1, 1, 1, 1, 4],
    [2, 1, -1, 1, 2],
    [1, -1, 2, -1, 2],
    [3, 1, 1, 2, 6]
])

print("Matriks Awal:")
print(B)

# Langkah 1: Eliminasi kolom pertama
M1 = B
M1.add_multiple_of_row(1, 0, -2)
M1.add_multiple_of_row(2, 0, -1)
M1.add_multiple_of_row(3, 0, -3)

print("\nLangkah 1:")
print(M1)

# -------------------------
# Langkah 2: Buat pivot kedua jadi 1
M2 = M1
M2.rescale_row(1, -1)

print("\nLangkah 2:")
print(M2)

# -------------------------
# Langkah 3: Eliminasi kolom kedua
M3 = M2
M3.add_multiple_of_row(2, 1, 2)
M3.add_multiple_of_row(3, 1, 2)

print("\nLangkah 3:")
print(M3)

# -------------------------
# Langkah 4: Buat pivot ketiga jadi 1
M4 = M3
M4.rescale_row(2, 1/7)

print("\nLangkah 4:")
print(M4)

# -------------------------
# Langkah 5: Eliminasi kolom ketiga
M5 = M4
M5.add_multiple_of_row(3, 2, -4)

print("\nLangkah 5 (Eselon Baris):")
print(M5)
```

# Eliminasi Gaussian (4 Variabel)

## Matriks Augmentasi Awal

$$
\begin{bmatrix}
1 & 1 & 1 & 1 & 4 \\
2 & 1 & -1 & 1 & 2 \\
1 & -1 & 2 & -1 & 2 \\
3 & 1 & 1 & 2 & 6
\end{bmatrix}
$$

---

## Langkah 1: Eliminasi Kolom Pertama

Operasi baris:

R2 = R2 - 2R1  
R3 = R3 - R1  
R4 = R4 - 3R1  

$$
\begin{bmatrix}
1 & 1 & 1 & 1 & 4 \\
0 & -1 & -3 & -1 & -6 \\
0 & -2 & 1 & -2 & -2 \\
0 & -2 & -2 & -1 & -6
\end{bmatrix}
$$

---

## Langkah 2: Membuat Pivot Kedua = 1

R2 = (-1) R2  

$$
\begin{bmatrix}
1 & 1 & 1 & 1 & 4 \\
0 & 1 & 3 & 1 & 6 \\
0 & -2 & 1 & -2 & -2 \\
0 & -2 & -2 & -1 & -6
\end{bmatrix}
$$

---

## Langkah 3: Eliminasi Kolom Kedua

R3 = R3 + 2R2  
R4 = R4 + 2R2  

$$
\begin{bmatrix}
1 & 1 & 1 & 1 & 4 \\
0 & 1 & 3 & 1 & 6 \\
0 & 0 & 7 & 0 & 10 \\
0 & 0 & 4 & 1 & 6
\end{bmatrix}
$$

---

## Langkah 4: Membuat Pivot Ketiga = 1

R3 = (1/7) R3  

$$
\begin{bmatrix}
1 & 1 & 1 & 1 & 4 \\
0 & 1 & 3 & 1 & 6 \\
0 & 0 & 1 & 0 & 10/7 \\
0 & 0 & 4 & 1 & 6
\end{bmatrix}
$$

---

## Langkah 5: Eliminasi Kolom Ketiga

R4 = R4 - 4R3  

$$
\begin{bmatrix}
1 & 1 & 1 & 1 & 4 \\
0 & 1 & 3 & 1 & 6 \\
0 & 0 & 1 & 0 & 10/7 \\
0 & 0 & 0 & 1 & 2/7
\end{bmatrix}
$$

---

## Bentuk Eselon Baris

Matriks sudah berbentuk eselon baris.

Dari sini dapat dilakukan substitusi balik untuk mendapatkan solusi.
